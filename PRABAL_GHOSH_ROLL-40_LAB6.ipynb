```python
# PRE PROCESSING FOR BagofWords

from nltk import sent_tokenize
import re
documents = sent_tokenize(result_text)
documents = [word.replace('\n','') for word in documents]
print(documents)
print("No of sentences: ",len(documents))
```

```python
#BagofWords
docV = []
for doc in documents:
    dt = np.zeros(len(vocabulary1))
    for w in doc.split():
        if w.lower() in vocabulary1:
            indexOfW = vocabulary1.index(w.lower())
            dt[indexOfW] +=1
    docV.append(dt)
```

```python
def cosSimilar(v1,v2): 
    sum = 0
    lv1 = 0
    lv2 = 0
    for i in range(v1.shape[0]):
        sum += v1[i]*v2[i]
        lv1 += v1[i]*v1[i]
        lv2 += v2[i]*v2[i]
    dp = sum / (math.sqrt(lv1) * math.sqrt(lv2))
    return dp
```

```python
def textToVector(a):
    vector = np.zeros(len(vocabulary1))
    for word in a.split():
        if word.lower() in vocabulary1:
            wordindex = vocabulary1.index(word.lower())
            vector[wordindex] +=1
        return vector
```
