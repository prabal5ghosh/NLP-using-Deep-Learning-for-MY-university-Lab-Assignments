```python
**Using the pretrained BERT model implement Question Answering System**
```

```python
import torch
```

```python
from transformers import BertForQuestionAnswering
```

```python
model=BertForQuestionAnswering.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
```

```python
from transformers import BertTokenizer
tokenizer=BertTokenizer.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
```

```python
question="what is a Maxwell's equations?"
paragraph="Maxwell's equations are a set of coupled partial differential equations that, together with the Lorentz force law, form the foundation of classical electromagnetism, classical optics, and electric circuits. The equations provide a mathematical model for electric, optical, and radio technologies, such as power generation, electric motors, wireless communication, lenses, radar etc. They describe how electric and magnetic fields are generated by charges, currents, and changes of the fields."
```

```python
question='[CLS] '+question+ '[SEP]'
paragraph=paragraph+ '[SEP]'
```

```python
question_tokens=tokenizer.tokenize(question)
paragraph_tokens=tokenizer.tokenize(paragraph)
```

```python
tokens=question_tokens+paragraph_tokens
input_ids=tokenizer.convert_tokens_to_ids(tokens)
```

```python
segment_ids=[0] * len(question_tokens)
segment_ids+=[1] * len(paragraph_tokens)
```

```python
input_ids=torch.tensor([input_ids])
segment_ids=torch.tensor([segment_ids])
```

```python
outputs=model(input_ids, token_type_ids= segment_ids)
```

```python
start_scores=outputs.start_logits
end_scores=outputs.end_logits
```

```python
start_answer=torch.argmax(start_scores)
end_answer=torch.argmax(end_scores)

answer=' '.join(tokens[start_answer:end_answer+1])
print('Answer:"'+answer+'"')
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

scibert_tokenizer = AutoTokenizer.from_pretrained("allenai/scibert_scivocab_uncased")
scibert_model = AutoModel.from_pretrained("allenai/scibert_scivocab_uncased")

print('scibert_tokenizer is type:', type(scibert_tokenizer))
print('    scibert_model is type:', type(scibert_model))
```

```python
from transformers import BertTokenizer, BertModel

bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')
```

```python
text = "BERT is a Bidirectional Encoder Representations from Transformers. It is one of the most popular and widely used NLP models. BERT models can consider the full context of a word by looking at the words that come before and after it, which is particularly useful for understanding the intent behind the query asked. Because of its bidirectionality, it has a deeper sense of language context and flow and hence, is used in a lot of NLP tasks nowadays. More details about BERT in the article along with the code."


bert_tokens = bert_tokenizer.tokenize(text)
scibert_tokens = scibert_tokenizer.tokenize(text)

while len(scibert_tokens) < len(bert_tokens):
    scibert_tokens.append("")


print('{:<12} {:<12}'.format("BERT", "SciBERT"))
print('{:<12} {:<12}'.format("----", "-------"))


for tup in zip(bert_tokens, scibert_tokens):
    print('{:<12} {:<12}'.format(tup[0], tup[1]))
```

```python
biobert_tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
biobert_model = AutoModel.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
```

```python
text = "A clinical biologist is a health professional such as a doctor of medicine, pharmacist, or biologist that is specialized in clinical biology, a medical specialty derived from clinical pathology. The concept includes interventional biology, including assisted reproductive technology."
text = '[CLS] ' + text + '[SEP]'
brt_tokens = bert_tokenizer.tokenize(text)
sci_tokens = scibert_tokenizer.tokenize(text)
bio_tokens = biobert_tokenizer.tokenize(text)
print(brt_tokens[:20], "number of tokens: ", len(brt_tokens))
print(sci_tokens[:20], "number of tokens: ", len(sci_tokens))
print(bio_tokens[:20], "number of tokens: ", len(bio_tokens))
```

```python
model = BertModel.from_pretrained('bert-base-uncased', output_hidden_states = True)
model.eval()
```

```python
def find_bert_vector(text):
    token = bert_tokenizer.tokenize(text)
    tid = tokenizer.convert_tokens_to_ids(token)
    ttensor = torch.tensor([tid])
    with torch.no_grad():
        outputs = model(ttensor)
        hidden_states = outputs[2]
    tembeddings = torch.stack(hidden_states, dim=0)
    tembeddings = torch.squeeze(tembeddings, dim=1)
    tembeddings = tembeddings.permute(1,0,2)
    print("Vector of ", text, ": ", torch.cat((tembeddings[0][-1], tembeddings[0][-2], tembeddings[0][-3], tembeddings[0][-4]), dim=0))
```

```python
find_bert_vector("hello")
```
